In February 2013 from shop 1, 1 product of  22154  'item_id' was  sold

write aprogramm to determine total count of unique 'item_id' or items are being sold permonth from a unique shop 
as example
item_id  shop_id  item_cnt_month
   3432        2               4

total 4 nubers of items haveing 3432 "item_id" was being sold per month from 'shop_id'2        

```python
df_pro=[]
df_c=[]

for i in np.arange(0,1509,1):
    df_c.append(df_final[i]['item_id'].value_counts())
    l1=list(df_c[i].index)
    l2=list(df_c[i])
    l3=df_final[i]['shop_id'][: len(df_c[i])]
    data = {'item_id': l1,
        'shop_id': l3,
           'item_cnt_month':l2}
    df_pro.append(pd.DataFrame(data))
    
    
```

```python
def sample(j):
    df_j=[]
    for i in np.arange(0,1505,1):
          df_j.append(df_for_train[i].sample(n = j, replace = True)) 
    return df_j   
```
